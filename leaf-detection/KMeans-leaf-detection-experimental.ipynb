```python
import numpy as np
from sklearn.cluster import KMeans
from scipy import ndimage
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
figsize(20, 12)
import matplotlib.image as mpimg

# img = mpimg.imread('../data/1/NC-1_2019-04-02_14-39-00.jpg')
img = mpimg.imread('../data/12/NC-12_2019-04-02_11-31-46.jpg') # load image
w = img.shape[0]
h = img.shape[1]

plt.imshow(img)

img = img.reshape(w*h, 3) # w*h inputs with 3 features
print(img.shape)

km = KMeans(n_clusters=3, random_state=0) # 3 clusters. Feel free to play around
# img must be a NxM 2d array where N is the number of inputs and M is the number of features
%time clusters = km.fit_predict(img) 
"""
clusters is a vector of length w*h where each value is a number from 0 to n-1 
with each field representing the cluster correspoding input was assigned to
"""
print(clusters.shape)
```

```python
c = clusters # set alias for clusters
# c = c/np.linalg.norm(c, 2) # normalize values for displaying image
fimg = c.reshape(w, h)
plt.imshow(fimg)
```
